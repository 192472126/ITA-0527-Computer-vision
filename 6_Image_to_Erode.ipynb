```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Read an image ---
# Replace 'path/to/your/image.jpg' with the actual path to your image file
# For example, if you upload an image named 'test_image.jpg' to your Colab environment,
# you can use 'test_image.jpg' as the path.
# If the image is not found, imread will return None.
image_path = 'image.jpg'
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not open or find the image at {image_path}")
else:
    # OpenCV reads images in BGR format by default, convert to RGB for matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # --- 2. Convert an Image to erode using Erode function ---
    # Define a kernel for erosion. A common kernel is a 5x5 matrix of ones.
    kernel = np.ones((5,5), np.uint8)

    # Apply the erode function
    eroded_image = cv2.erode(image, kernel, iterations=1)
    eroded_image_rgb = cv2.cvtColor(eroded_image, cv2.COLOR_BGR2RGB)

    # --- Display the original and eroded images ---
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(eroded_image_rgb)
    plt.title('Eroded Image')
    plt.axis('off')

    plt.show()
```
