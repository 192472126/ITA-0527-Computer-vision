```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the uploaded image
img_original = Image.open('/content/image.jpg')

# Perform a 270-degree clockwise rotation
# In PIL's Image.rotate method, a positive angle is counter-clockwise.
# So, a 270-degree clockwise rotation is equivalent to a 90-degree counter-clockwise rotation.
img_rotated = img_original.rotate(angle=90, expand=True)

# Display the original and rotated images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_original)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_rotated)
plt.title('270° Clockwise Rotated Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
