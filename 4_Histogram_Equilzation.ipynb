```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# --- Image Upload Section (Optional, if you haven't uploaded manually) ---
# If you prefer to upload via code, uncomment and run this section.
# uploaded = files.upload()
# for filename in uploaded.keys():
#   image_path = filename
#   print(f"Uploaded file: {image_path}")
#   break # Assuming only one file is uploaded

# --- Set your image path here ---
# Replace 'your_image.jpg' with the actual name of your uploaded image file.
image_path = 'image.jpg'  # <--- IMPORTANT: Changed to 'image.jpg'

# Load the image in grayscale
# Using cv2.IMREAD_GRAYSCALE ensures the image is loaded as a single channel
img_original = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img_original is None:
    print(f"Error: Could not load image from {image_path}. Please check the path and if the file exists.")
else:
    # Apply histogram equalization
    img_equalized = cv2.equalizeHist(img_original)

    # Plotting
    plt.figure(figsize=(15, 6))

    # Original Image
    plt.subplot(2, 2, 1)
    plt.imshow(img_original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Histogram of Original Image
    plt.subplot(2, 2, 2)
    plt.hist(img_original.flatten(), 256, range=[0, 256], color='r') # Updated range parameter
    plt.title('Histogram of Original Image')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    # Equalized Image
    plt.subplot(2, 2, 3)
    plt.imshow(img_equalized, cmap='gray')
    plt.title('Equalized Image')
    plt.axis('off')

    # Histogram of Equalized Image
    plt.subplot(2, 2, 4)
    plt.hist(img_equalized.flatten(), 256, range=[0, 256], color='r') # Updated range parameter
    plt.title('Histogram of Equalized Image')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```
