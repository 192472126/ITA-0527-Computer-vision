```python
from PIL import Image
import matplotlib.pyplot as plt

# --- Instructions ---
# 1. Upload an image to your Colab environment or provide a valid path.
#    For example, if you upload `my_image.jpg` to the files section (left panel),
#    you can use 'my_image.jpg' as the image_path.
# 2. Replace 'path/to/your/image.jpg' with the actual path to your image file.
# --------------------

image_path = 'image.jpg'  # <--- REPLACE THIS WITH YOUR IMAGE PATH

try:
    # Load the image
    original_image = Image.open(image_path)

    # Perform a 180-degree rotation along the y-axis (horizontal flip)
    # In PIL, Image.FLIP_LEFT_RIGHT performs a horizontal flip.
    rotated_image = original_image.transpose(Image.FLIP_LEFT_RIGHT)

    # Display the original and rotated images
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(original_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(rotated_image)
    axes[1].set_title('180-degree Y-axis Rotation (Horizontal Flip)')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    print(f"Original image size: {original_image.size}")
    print(f"Rotated image size: {rotated_image.size}")

    # Optional: Save the rotated image
    # rotated_image.save('rotated_image_y_axis.jpg')
    # print("Rotated image saved as 'rotated_image_y_axis.jpg'")

except FileNotFoundError:
    print(f"Error: Image file not found at '{image_path}'. Please ensure the path is correct and the file exists.")
except Exception as e:
    print(f"An error occurred: {e}")
```
