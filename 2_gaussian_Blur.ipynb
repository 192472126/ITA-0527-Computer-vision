```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Read the image
# Replace '/path/to/your/image.jpg' with the actual path to your image file.
# You might need to upload an image to your Colab environment or specify a URL.
try:
    # Example: If you have an image named 'my_image.jpg' in your Colab environment
    image_path = 'image.jpg' # <<<--- IMPORTANT: Change this to your image path
    original_image = cv2.imread(image_path)

    if original_image is None:
        raise FileNotFoundError(f"Error: Image not found at {image_path}. Please check the path.")

    # OpenCV reads images in BGR format, matplotlib expects RGB
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # Step 2: Convert the Image to Blur using Gaussian Blur
    # Define the kernel size for Gaussian blur. It must be a positive odd integer.
    # (width, height) tuple, e.g., (5, 5), (9, 9), etc.
    kernel_size = (15, 15)
    # Standard deviation in X and Y direction. If 0, it's calculated from kernel size.
    sigma_x = 0
    sigma_y = 0

    blurred_image = cv2.GaussianBlur(original_image_rgb, kernel_size, sigma_x, sigma_y)

    # Step 3: Display the original and blurred images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off') # Hide axes for cleaner image display

    plt.subplot(1, 2, 2)
    plt.imshow(blurred_image)
    plt.title(f'Blurred Image (Gaussian Blur with kernel={kernel_size})')
    plt.axis('off')

    plt.show()

except FileNotFoundError as e:
    print(e)
    print("Please upload an image file (e.g., JPG, PNG) to your Colab environment and update 'image_path' with its name or full path.")
except Exception as e:
    print(f"An error occurred: {e}")
```
