```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# --- Image Loading ---
# Replace 'path/to/your/image.jpg' with the actual path to your image file.
# If the image is not found, a dummy image will be created for demonstration.
image_path = '/content/image.jpg'
try:
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Image not found at {image_path}. Creating a dummy image.")
except FileNotFoundError:
    print(f"Warning: {image_path} not found. Creating a dummy image for demonstration.\n")
    # Create a simple dummy image (e.g., a square with a gradient and text)
    dummy_width, dummy_height = 300, 200
    img = np.zeros((dummy_height, dummy_width, 3), dtype=np.uint8)
    # Add a horizontal gradient (red channel)
    img[:, :, 2] = np.linspace(0, 255, dummy_width, dtype=np.uint8)[np.newaxis, :]
    # Add a vertical gradient (green channel)
    img[:, :, 1] = np.linspace(0, 255, dummy_height, dtype=np.uint8)[:, np.newaxis]
    cv2.putText(img, "Original Image", (20, 100), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

# OpenCV reads images in BGR format, convert to RGB for matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# --- Perform 90-degree clockwise rotation ---
# This performs a standard 2D rotation of the image content 90 degrees clockwise.
# The top edge moves to the right, the right edge to the bottom, and so on.
rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
rotated_img_rgb = cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB)

# --- Display Images ---
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off') # Hide axes for cleaner image display

plt.subplot(1, 2, 2)
plt.imshow(rotated_img_rgb)
plt.title('Rotated 90 Degrees Clockwise')
plt.axis('off') # Hide axes for cleaner image display

plt.tight_layout() # Adjust layout to prevent titles/labels from overlapping
plt.show()
```
